```python
import numpy as np


#input matrix
A= np.array([[1,0],[0,1],[1,1]],dtype=float)

#compute A^T A
ATA= A.T @ A

#find eigen value and eigen decompositon

eigen_value,v= np.linalg.eig(ATA)

# sort the eigen value and eigen vector in deceding order

idx= np.argsort(-eigen_value)
eigen_value=eigen_value[idx]

#compute singular value
v= v[:,idx]

singular_values= np.sqrt(eigen_value)

# compute left singular value vecotr

U= np.zeros((A.shape[0],A.shape[0]))

for i in range(len(singular_values)):
    U[:,i]= (A @ v[:,i])/singular_values[i]

#normalize the column of U

U[:,:len(singular_values)] /= np.linalg.norm(U[:,:len(singular_values)],axis=0)

# construct sigma

Sigma= np.zeros_like(A,dtype=float)
np.fill_diagonal(Sigma,singular_values)

# reconstruct A 

A_reconstruct= U[:,:len(singular_values)] @ Sigma @ v.T

print("A =\n", A)
print("\nA^T A =\n", ATA)
print("\nEigenvalues of A^T A:\n", eigen_value)
print("\nSingular values:\n", singular_values)
print("\nRight singular vectors (V):\n", v)
print("\nLeft singular vectors (U):\n", U)
print("\nÎ£ (Sigma):\n", Sigma)
print("\nReconstructed A =\n", A_reconstruct)





```
