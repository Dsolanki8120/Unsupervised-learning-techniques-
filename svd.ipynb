```python
import numpy as np

# Input matrix
A = np.array([[1, 0],
              [0, 1],
              [1, 1]], dtype=float)

# Compute A^T A
ATA = A.T @ A

# Eigen decomposition
eigen_value, v = np.linalg.eig(ATA)

# Sort in descending order
idx = np.argsort(-eigen_value)
eigen_value = eigen_value[idx]
v = v[:, idx]

# Singular values
singular_values = np.sqrt(eigen_value)

# Left singular vectors U = A * v / σ
U = np.zeros((A.shape[0], len(singular_values)))

for i in range(len(singular_values)):
    U[:, i] = (A @ v[:, i]) / singular_values[i]

# Normalize U columns
U = U / np.linalg.norm(U, axis=0)

# Construct Σ (2×2 because rank = 2)
Sigma = np.diag(singular_values)

# Reconstruct A
A_reconstruct = U @ Sigma @ v.T

print("A =\n", A)
print("\nA^T A =\n", ATA)
print("\nEigenvalues of A^T A:\n", eigen_value)
print("\nSingular values:\n", singular_values)
print("\nRight singular vectors (V):\n", v)
print("\nLeft singular vectors (U):\n", U)
print("\nΣ (Sigma):\n", Sigma)
print("\nReconstructed A =\n", A_reconstruct)
```
